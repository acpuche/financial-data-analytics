```python
# Se importan las librerías 

import numpy as np
import pandas as pd
```

```python
# Se carga la base de datos, se convierte de csv a un dataframe de pandas

dataset = pd.read_csv('/workspaces/financial-data-analysis/data/Coca_cola_prices.csv')
dataset.head(5)
```

```python
# Calculo del valor promedio y la desviación estándar a la columna precio

price_mean = np.mean(dataset['Price'])
print(f'The mean price is: {price_mean}')

price_std = np.std(dataset['Price'])
print(f'The standar deviation price is: {price_std}')
```

```python
# Se aplica la fórmula rt=(Pt - Pt-1)/Pt-1

dataset['Returns'] = (dataset['Price']-dataset['Price'].shift(-1))/dataset['Price'].shift(-1)
dataset.head(5)
```

```python
# Se aplica la fórmula logarítmica: rt = ln(Pt) - ln(Pt-1)

dataset['Returns(log)'] = (np.log(dataset['Price'])-np.log(dataset['Price'].shift(-1)))

dataset.head(5)
```
